```python
import json

def menu():
    print("===================")
    print("1. Create Student ")
    print("2. Update Student ")
    print("3. Delete Student ")
    print("4. Search Student ")
    print("5. Exit ")
    print("===================")
    

class student():
    #with open("student.json","w") as f:
    #    json.dump({},f);
        
    def __init__(self,name="",fatherName="",classEnrolled="",rollNumber="",subjectList=""):
        self.Name=name;
        self.FatherName=fatherName;
        self.ClassName=classEnrolled;
        self.RollNumber=rollNumber;
        self.SubjectList=subjectList;
    
    @staticmethod
    def student_exist(rollNumber):
        with open("student.json") as f:
            temp_dict = {k:v for(k,v) in json.load(f).items() if v["rollNumber"]==rollNumber}
            if(len(temp_dict)>0):
                return True
            else:
                return False        
            
    def __find_student(self,dictArguments,axactSearch=0):
        with open("student.json") as f:
            #rcheck = self.RollNumber!=""?v["rollNumber"]==self.RollNumber:true;
            if(axactSearch==0):
                temp_dict = {k:v for(k,v) in json.load(f).items() if v["rollNumber"]==dictArguments["rollNumber"] or v["name"]==dictArguments["name"] or v["fName"]==dictArguments["fatherName"] or v["class"]==dictArguments["className"]}
            else:
                temp_dict = {k:v for(k,v) in json.load(f).items() if all([dictArguments["rollNumber"]!="" and v["rollNumber"]==dictArguments["rollNumber"],dictArguments["name"]!="" and v["name"]==dictArguments["name"],dictArguments["fatherName"]!="" and v["fName"]==dictArguments["fatherName"],dictArguments["className"]!="" and v["class"]==dictArguments["className"]])}
             
            if(len(temp_dict)>0):
                return {"isValid":True,"message":"Student Found","result":temp_dict}
            else:
                return {"isValid":False,"message":"No Student Found","result":temp_dict}
            
        
    def update_student(self):
        if(student.student_exist(self.RollNumber)):
            student.delete_student(self.RollNumber)
            self.create_student()
            return {"isError":False,'message':"Student Record Update Successfully"}
        else:
            return {"isError":True,'message':"Student Record Not Found"}            
    
    @staticmethod
    def delete_student(rollNumber):
        stu = student.GetStudent(rollNumber)
        if(stu["isExist"]):
            with open("student.json","r") as f:
                x = json.load(f)
                del x[rollNumber]
            with open("student.json","w") as f:
                json.dump(x,f)
                return {"isError":False,'message':"Student Record Deleted Successfully"}
        else:
            return {"isError":True,'message':"Student Record Not Found"}
            
    
    def create_student(self):
        if(student.student_exist(self.RollNumber)):
            return {"isError":True,'message':"Error:Student Already Exist, Please Enter Different Roll Number"}
        else:
            studentDict={}
            stuDict={}
            
            with open("student.json","r") as f:
                x = json.load(f)
                stuDict = x

            studentDict["name"] = self.Name
            studentDict["fName"] = self.FatherName
            studentDict["class"] = self.ClassName
            studentDict["rollNumber"] = self.RollNumber
            studentDict["subject"] = self.SubjectList
            stuDict[stu.RollNumber] = studentDict

            with open("student.json","w") as f:
                json.dump(stuDict,f)
                return {"isError":False,'message':"Student Created Successfully"}
    
    def search_student(self,name="",fatherName="",className="",rollNumber="",axactSearch=0):
        dictArguments = {'name':name,'fatherName':fatherName,'className':className,'rollNumber':rollNumber}
        stu_dict = self.__find_student(dictArguments,axactSearch)
        if(stu_dict["isValid"]):
            for s in stu_dict["result"].values():
                print("Name        : ",s["name"])
                print("Father Name : ",s["fName"])
                print("Class  Name : ",s["class"])
                print("Roll Number : ",s["rollNumber"])
                print("Subjects ")
                for i in s["subject"]:
                    print("\t",end="")
                    print(i)
                print("======================================")

    @staticmethod
    def GetStudent(rollNumber):
        with open("student.json") as f:
            temp_dict = {k:v for(k,v) in json.load(f).items() if v["rollNumber"]==rollNumber}
            if(len(temp_dict)>0):
                stu = student(temp_dict[rollNumber]["name"],temp_dict[rollNumber]["fName"],temp_dict[rollNumber]["class"],temp_dict[rollNumber]["rollNumber"],temp_dict[rollNumber]["subject"])
                return {"isExist":True,"message":"Student Record Found","objvalue":stu}
            else:
                stu = student()
                return {"isExist":False,"message":"Student Record Not Found","objvalue":stu}
                
            #if(len(temp_dict)>0):
            #    return {"isError":0,"message":"Record Found","objValue":temp_dict}
            #else:
            #    return {"isError":1,"message":"Record Not Found","objValue":temp_dict}
            
            
        
    #def fromBirthYear(cls, name, year): 
    #    return cls(name, date.today().year - year) 
```

```python
flag = True
while(flag):
    menu()
    option = int(input(""))
    if(option==5):
        flag=False
    elif(option==1):
        addMoreStudent = True
        while(addMoreStudent):
            #print("\n")
            validRollNumber=True
            while(validRollNumber):
                rollNumber = input("Enter Roll Number : ")
                validRollNumber = student.student_exist(rollNumber)
                if(validRollNumber is True):
                    print("Roll Number Already exist, Please Enter new Roll Number : ")
                
            name = input("Enter Student Name : ")
            fatherName = input("Enter Student Father Name : ")
            className = input("Enter Class Name : ")
            
            subjects = []
            #print("\n")
            print("\nEnter Subject. Type Q for quit\n")
            #print("\n")
            subjectEnd = True
            while(subjectEnd):
                sub = input("Enter Subject Name : ")
                if(sub.lower()=='q'):
                    subjectEnd=False
                else:
                    subjects.append(sub)
            stu = student(name,fatherName,className,rollNumber,subjects)
            result = stu.create_student()
            print("\n -------------------")
            print(result["message"])
            print("-------------------\n")
            if(result["isError"] is not True):
                opt = input("Do you want to add more student (Y/N) : ")
                if(opt.lower()=='n'):
                    addMoreStudent = False
    elif(option==2):
        rollNumber = input("Enter Roll Number : ")
        obj = student.GetStudent(rollNumber)
        if(obj["isExist"]):
            stu = obj["objvalue"]
            stu.Name = input("Enter Student Name : ")
            stu.FatherName = input("Enter Student Father Name : ")
            stu.ClassName = input("Enter Student Class Name : ")
            stu.update_student()
        else:
            print(obj["message"])
        
                
                
        
    

```
