```python
import os
import numpy as np
from PIL import Image as image

def imageToNp(fileName):
    try:
        npArray = np.array(image.open(fileName).resize((200,200)))
        return npArray
    except Exception as e:
        print("Error")
        

dir="photos/"

files = next(os.walk(dir))[2]

file_count = len(files)
imageArray = np.zeros([file_count,200,200,3])

for i,file in enumerate(files):
    imageArray[i] = imageToNp(dir+file)
    
for i in range(len(imageArray)):
    print(imageArray[i])
    print("+================")
    

    
```

```python
import pandas as pd

apples = pd.Series([3,2,0,1])
oranges = pd.Series([3,4,7,8])
list = [3,4,5,6]

data = {"apple":apples,"oranges":oranges,"banana":list}
fruits_df = pd.DataFrame(data)
print(fruits_df)
```

```python
import pandas as pd

apples = pd.Series([3,2,0,1],['a','b','c','d'])
oranges = pd.Series([3,4,7,8],index=['e','f','g','h'])
#list = [3,4,5,6]

data = {"apple":apples,"oranges":oranges}
fruits_df = pd.DataFrame(data)
print(fruits_df)
```

```python
import pandas as pd

apples = pd.Series([3,2,0,1],['a','b','c','d'],name="Apples")
oranges = pd.Series([3,4,7,8],index=['e','f','g','h'],name="Oranges")
oranges.index = apples.index
#list = [3,4,5,6]

data = {"apple":apples,"oranges":oranges}
#data = [apples,oranges]
fruits_df = pd.DataFrame(data)
print(fruits_df)
```
