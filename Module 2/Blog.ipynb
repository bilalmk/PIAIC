# Pandas.cut to binning or bucketing the data

```python
op_labels = ['Shety','Moderate','Good']
category = [0.,4.,7.,10.]

movies_df['imdb_labels'] = pd.cut(movies_df['imdb_score'],bins=category,labels=op_labels,right=True,include_lowest=True)
movies_df[['movie_title','imdb_score','imdb_labels','facenumber_in_poster']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

movies_df.boxplot(column=["facenumber_in_poster"],grid=True,return_type='axes')
```
